```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv(r'C:\Users\ALbayan\Desktop\AS_Tm\price_car\train_set_.csv')
df.head()
```

```python
df[['ID', 'Name', 'Location', 'Year', 'Kilometers_Driven', 'Fuel_Type', 'Transmission', 'Owner_Type', 'Mileage', 'Engine', 'Power', 'Seats', 'Price']] = df['ID;Name;Location;Year;Kilometers_Driven;Fuel_Type;Transmission;Owner_Type;Mileage;Engine;Power;Seats;Price'].str.split(';', expand=True)
df.drop('ID;Name;Location;Year;Kilometers_Driven;Fuel_Type;Transmission;Owner_Type;Mileage;Engine;Power;Seats;Price', axis=1, inplace=True)
df.head()
df.info
```

```python
df_without_ID=df.drop('ID', axis=1)
df_without_ID.head()
```

```python
df_without_ID['Mileage'] = pd.to_numeric(
    df_without_ID['Mileage'].astype(str).str.replace(r'[^0-9.]', '', regex=True),
    errors='coerce'
)

df_without_ID['Engine'] = pd.to_numeric(
    df_without_ID['Engine'].astype(str).str.replace(r'[^0-9.]', '', regex=True),
    errors='coerce'
)

df_without_ID['Power'] = pd.to_numeric(
    df_without_ID['Power'].astype(str).str.replace(r'[^0-9.]', '', regex=True),
    errors='coerce'
)
df_without_ID['Seats'] = pd.to_numeric(
    df_without_ID['Seats'].astype(str).str.replace(r'[^0-9.]', '', regex=True),
    errors='coerce'
)
df_without_ID['Seats'] .astype('float64').fillna(df_without_ID['Seats'].mean(), inplace=True)
df_without_ID['Year'] = pd.to_numeric(df_without_ID['Year'], errors='coerce').astype('float64')
df_without_ID['Kilometers_Driven'] = pd.to_numeric(df_without_ID['Kilometers_Driven'], errors='coerce').astype('float64')
df_without_ID.info()
df_without_ID.head()
```
